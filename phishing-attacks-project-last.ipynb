```python
# "Domain" sütununun veri setinden çıkarılması
data = data0.drop(['Domain'], axis = 1).copy()
```

```python
# "Domain" sütununun çıkarılmasıyla 16 feature kaldı
# "URL Derinliği" veri setinde max değer 20 olduğu için veri setinden çıkarmıyoruz
# Çok geniş bir değer kümesi aralığına sahip olmadığı için sorun yaratmayacak
```

```python
# pre-processing aşamasının tamamlanması için verileri kontrol ediyoruz
data.isnull().sum()
```

```python
# veri setinde ilk 5000 değer yasal, son 5000 değer ise phishing olacak şekilde
# training ve test şeklinde veri setini böleceğimizde yığılma yaşanmasını engellemek için
# veri setini karıştırmamız gerekiyor. over-training'i ortadan kaldırmak için bunu uygulamalıyız
```

```python
# verilerin karıştırılması
data = data.sample(frac=1).reset_index(drop=True)
data.head()
```

```python
# özelliklerin ve hedef olan sütunların X ve Y şeklinde ayrılması ve "Label" atanması
y = data['Label']
X = data.drop('Label',axis=1)
X.shape, y.shape
```

```python
# veri setinin training ve test olacak şekilde 80/20 oranında bölünmesi
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 12)
X_train.shape, X_test.shape
```

```python
# buradan sonrasında Machine Learning modellerini kullanıcaz
# model oluşturup eğitim yaparak sayısal veriler elde edicez
# başlangıç olarak Decision Tree ve Random Forest kullanıcaz
# çünkü bu modelleri daha önce "Machine Learning" dersinde kullandık
```

```python
# paketlerin ve kütüphanelerin yüklenmesi
from sklearn.metrics import accuracy_score
```

```python
# sonuçları saklamak için temp oluşturma (model performansı)
ML_Model = []
acc_train = []
acc_test = []

# sonuçları yazdıracak olan fonksiyon
def storeResults(model, a,b):
  ML_Model.append(model)
  acc_train.append(round(a, 3))
  acc_test.append(round(b, 3))
```

```python
# 1. MODEL: Decision Tree Classifier
# Karar ağacı modeli geçen dönem kullandığımız bir model
# sınıflandırma ve regresyon operasyonları için kullanıldığını biliyoruz
# Python üzerinden anlatılacak olursa, if/else yöntemiyle sınıflandırma yapıyor diyebiliriz
# bir ağaç oluşturup, yukarıdan aşağıya doğru seçimler yaparak bize sonuçları verecek
```

```python
# Decision Tree modelinin yüklenmesi
from sklearn.tree import DecisionTreeClassifier

# modeli somutlaştırma ve derinlik atama
tree = DecisionTreeClassifier(max_depth = 5)

# modelin fit değerlerini ver
tree.fit(X_train, y_train)
```

```python
# örnekler için modelden target değeri tahmini oluşturma
y_test_tree = tree.predict(X_test)
y_train_tree = tree.predict(X_train)
```

```python
# Decision Tree Performans Değerlendirmesi
```

```python
# model performansının hesaplanması
acc_train_tree = accuracy_score(y_train,y_train_tree)
acc_test_tree = accuracy_score(y_test,y_test_tree)

print("Decision Tree: Accuracy on training Data: {:.3f}".format(acc_train_tree))
print("Decision Tree: Accuracy on test Data: {:.3f}".format(acc_test_tree))
```

```python
# modeldeki feature'ların öneminin kontrolü
plt.figure(figsize=(9,7))
n_features = X_train.shape[1]
plt.barh(range(n_features), tree.feature_importances_, align='center')
plt.yticks(np.arange(n_features), X_train.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
# sonuçların depolanması, parametre sıralaması önemli
# duplicate olmasını engellemek için sadece 1 defa çalıştır
storeResults('Decision Tree', acc_train_tree, acc_test_tree)
```

```python
# 2. MODEL: Random Forest Classifier
# Random Forest modeli yine aynı şekilde sınıflandırma ve regresyonda kullanılıyor
# rastgele bir orman düşünün, her bir ağacın tahmin etme konusunda payı var
# ama büyük ihtimalle bu durum verilerin bir kısmında over-training'e yol açacaktır
# hepsi yüksek performanslı çalışan ve farklı şekillerde over-training'e yakın bir şekilde
# doğru tahminler yapan çok sayıda ağaç oluşturabilirsek over-training miktarını azaltabiliriz
```

```python
# Random Forest modelinin yüklenmesi
from sklearn.ensemble import RandomForestClassifier

# modeli somutlaştırma ve derinlik atama
forest = RandomForestClassifier(max_depth=5)

# modelin fit değerlerini ver
forest.fit(X_train, y_train)
```

```python
# Random Forest Performans Değerlendirmesi
```

```python
# model performansının hesaplanması
acc_train_forest = accuracy_score(y_train,y_train_forest)
acc_test_forest = accuracy_score(y_test,y_test_forest)

print("Random forest: Accuracy on training Data: {:.3f}".format(acc_train_forest))
print("Random forest: Accuracy on test Data: {:.3f}".format(acc_test_forest))
```

```python
# modeldeki feature'ların öneminin kontrolü
plt.figure(figsize=(9,7))
n_features = X_train.shape[1]
plt.barh(range(n_features), forest.feature_importances_, align='center')
plt.yticks(np.arange(n_features), X_train.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
# sonuçların depolanması, parametre sıralaması önemli
# duplicate olmasını engellemek için sadece 1 defa çalıştır
storeResults('Random Forest', acc_train_forest, acc_test_forest)
```

```python
# 3. MODEL: Multilayer Perceptrons (deep-learning)
# çok katmanlı algılayıcılar (MLP'ler), aynı zamanda ileri beslemeli sinir ağları
# hem sınıflandırma hem de regresyon problemlerinde uygulanabilir
# bir karara varmak için birden fazla işlem aşamasını gerçekleştiren doğrusal model
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.metrics import accuracy_score
import pandas as pd

# kullanmakta olduğumuz modeller
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# verileri yükle ve hazırla, X = feature_matrix, y = target_vector

models = {
    "SVM": SVC(),
    "MLP": MLPClassifier(max_iter=500, learning_rate_init=0.001),
    "XGBoost": XGBClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier()
}

results = []

kf = KFold(n_splits=5, shuffle=True, random_state=42)

for model_name, model in models.items():
    cv_scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
    results.append({
        "Model": model_name,
        "Accuracy Mean": cv_scores.mean(),
        "Accuracy Std": cv_scores.std()
    })

results_df = pd.DataFrame(results)
(results_df)
```

```python
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import warnings
from sklearn.exceptions import ConvergenceWarning

warnings.filterwarnings("ignore", category=ConvergenceWarning)

# kullanmakta olduğumuz modeller
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

models = {
    "SVM": SVC(),
    "MLP": MLPClassifier(max_iter=500, learning_rate_init=0.001),
    "XGBoost": XGBClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Decision Tree": DecisionTreeClassifier()
}

kf = KFold(n_splits=5, shuffle=True, random_state=42)
results = []

for model_name, model in models.items():
    train_accuracies = []
    test_accuracies = []
    
    for train_index, test_index in kf.split(X_new):
        X_train, X_test = X_new[train_index], X_new[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        model.fit(X_train, y_train)
        
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        
        train_accuracy = accuracy_score(y_train, train_pred)
        test_accuracy = accuracy_score(y_test, test_pred)
        
        train_accuracies.append(train_accuracy)
        test_accuracies.append(test_accuracy)
    
    results.append({
        "Model": model_name,
        "Train Accuracy Mean": np.mean(train_accuracies),
        "Train Accuracy Std": np.std(train_accuracies),
        "Test Accuracy Mean": np.mean(test_accuracies),
        "Test Accuracy Std": np.std(test_accuracies)
    })

results_df = pd.DataFrame(results)
(results_df)
```
